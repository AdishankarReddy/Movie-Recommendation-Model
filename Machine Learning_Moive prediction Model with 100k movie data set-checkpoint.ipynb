# Prediction Funtion

```python

def predict_movie(movie_name):
    movie_userwise_rating=movie_matrix[movie_name]
    similar_relation_other=movie_matrix.corrwith(movie_userwise_rating)
    similar_relation_other=similar_relation_other.dropna()
    corre_movie=pd.DataFrame(similar_relation_other,columns=['correlation with angry_movie'])
    
    corre_movies=corre_movie.join(rating['count_rating'])
    predictions=corre_movies[corre_movies['count_rating']>100].sort_values('correlation with angry_movie',ascending=False)
    return predictions
```

```python
predictions=predict_movie('Titanic (1997)')
```
